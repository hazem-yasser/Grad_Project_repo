```python
import numpy as np
import os

# --- CONFIGURATION ---
INPUT_DIR = "sim_data"   # Where your .txt files are
OUTPUT_DIR = "hex_files" # Where .hex files will go

print(f"--- CONVERTING {INPUT_DIR} TO HEX ---")

def txt_to_hex(filename):
    # 1. Read the Decimal Integers from .txt
    try:
        data = np.loadtxt(f"{INPUT_DIR}/{filename}", dtype=int)
    except OSError:
        print(f"[SKIP] Could not find {filename}")
        return

    # 2. Convert to 16-bit Hex Strings
    hex_lines = []
    # Handle both Arrays (Weights) and Vectors (Biases/Stimuli)
    iterator = data.flatten()
    
    for val in iterator:
        # 16-bit Two's Complement Logic
        val = int(val)
        if val < 0: 
            val += 65536
        hex_lines.append(f"{val:04x}")

    # 3. Save to .hex
    new_name = filename.replace(".txt", ".hex")
    with open(f"{OUTPUT_DIR}/{new_name}", "w") as f:
        f.write("\n".join(hex_lines))
    
    print(f" -> Converted {filename} to {OUTPUT_DIR}/{new_name}")

# --- LIST OF FILES TO CONVERT ---
# Must match the filenames in your 'sim_data' folder exactly
files_to_convert = [
    "L1_W.txt", "L1_b.txt",
    "L2_W.txt", "L2_b.txt",
    "L3_W.txt", "L3_b.txt",
    "input_stimuli.txt" # This is your X_test input
]

for f in files_to_convert:
    txt_to_hex(f)

print("\n[DONE] Hex files are ready in 'hex_files/'. Use these for SystemVerilog.")
```
